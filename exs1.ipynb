```python
df = pd.DataFrame({
'Name':['A','B','C','D','E'],
'Marks':[90,92,85,85,91]

})

df
```

```python
df['Rank'] = df['Marks'].rank(method='dense',ascending=True)
df
```

```python
df = pd.DataFrame({
'Name': ['A', 'B', 'C'],
'Maths': [90, 85, 95],
'Science': [88, 92, 90],
'English': [75, 80, 95]

})

df
```

```python
df[['Maths+rank','Science+rank','English+rank']] = df[['Maths','Science','English']].rank(ascending=True,method='dense')
df
```

```python
df['Total'] = df.apply(lambda x:x[1]+x[2]+x[3],axis=1)
df
```

```python
df['percetnilte'] = df['Total'].rank(pct=True)*100
df
```

```python


df['grade'] = np.where(
    df['percetnilte'] >= 75, "great",
    np.where(df['percetnilte'] >= 30, "good", "not good")
)
df
```
